```python
from youtube_transcript_api import YouTubeTranscriptApi
video_id = '7ARBJQn6QkM'
video_transcript=[]

try:
    yt_api = YouTubeTranscriptApi()
    transcript = yt_api.fetch(video_id)
    # transcript_text = ' '.join([item['text'] for item in transcript])
    for e in transcript:
        # print(e.text,'\n')
        video_transcript = ' '.join([e.text for e in transcript])

    print(video_transcript)

except Exception as e:
    print(f"An error occurred: {e}")
```

```python
from langchain_text_splitters import RecursiveCharacterTextSplitter

text_splitter = RecursiveCharacterTextSplitter(chunk_size=2000, chunk_overlap=700)
chunks = text_splitter.create_documents([video_transcript])

print(f'Created {len(chunks)} chunks.')
```

# Embedding Generation and Storing in Vector Store

```python
from langchain_google_genai import GoogleGenerativeAIEmbeddings
from langchain_community.vectorstores import FAISS

embed_model = GoogleGenerativeAIEmbeddings(model='models/gemini-embedding-001')
vector_store = FAISS.from_documents(chunks, embed_model)
```

```python
print(vector_store.index.ntotal == len(chunks))
(vector_store.index.ntotal, len(chunks))
```

```python
vector_store.index_to_docstore_id
```

```python
query = 'What is the vision for the future of robotics and physical AI?'
doc = vector_store.similarity_search(query=query, k=2)
for e in doc:
  print(e.page_content,'\n\n')
```

```python
vector_store.get_by_ids(['3dd155d2-14ed-4b53-a056-a87e174ba785'])
```

## Step 1 - Retrieval

```python
retriever = vector_store.as_retriever(search='similarity_search', search_kwargs={'k': 2})
```

```python
retriever
```

```python
retriever.invoke(query)
```

## Step 2 - Augmentation

```python
from langchain_google_genai import ChatGoogleGenerativeAI
chat_model = ChatGoogleGenerativeAI(model='gemini-2.5-flash', temperature=0.2)
```

```python
from langchain_core.prompts import PromptTemplate

prompt = PromptTemplate(
    template="""
      You are a helpful assistant.
      Answer ONLY from the provided transcript context.
      If the context is insufficient, just say you don't know.

      {context}
      Question: {question}
    """,
    input_variables = ['context', 'question']
)
```

```python
question = "is the topic of nuclear fusion discussed in this video? if yes then what was discussed"
retrieved_docs = retriever.invoke(question)
```

```python
context_txt = '\n\n'.join(e.page_content for e in retrieved_docs)
print(context_txt)
```

## Step 3 - Generation

```python
final_prompt = prompt.invoke({'context':context_txt, 'question':question})
print(final_prompt)
```

```python
response = chat_model.invoke(final_prompt)
response.content
```

## Building the same workflow in a form of Chain - LangChain's Advantage

```python
from langchain_core.output_parsers import StrOutputParser
parser = StrOutputParser()

def format_docs(docs):
  return '\n\n'.join(doc.page_content for doc in docs)
```

```python
from langchain_core.runnables import RunnablePassthrough, RunnableLambda, RunnableParallel
parallel_chain = RunnableParallel({'context': retriever | RunnableLambda(format_docs),
                          'question': RunnablePassthrough()})
chain = parallel_chain | prompt | chat_model | parser
```

```python
print(chain.invoke('What are the upcoming Technologies in the domain of AI'))
```
